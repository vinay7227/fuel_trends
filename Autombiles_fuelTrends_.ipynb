```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_types = pd.read_csv("C:/Users/vingolla/Downloads/stacked_area_chart_data.csv")
```

```python
df_fuel_types
```

```python
colors = ["#011638", "#7e2987", "ef2026"]
labels = ["Diesel", "Petrol", "Gas"]
sns.set_style("white")
plt.figure(figsize=(12, 6))
plt.stackplot(df_fuel_types["Year"],
             df_fuel_types["Diesel"],
             df_fuel_types["Petrol"],
             df_fuel_types["Gas"]
             )
plt.xticks(df_fuel_types["Year"], rotation = 45)
plt.legend(labels = labels, loc = "upper left")
plt.ylabel("Number of cars", fontsize =13)
plt.title("trend in engine fuels(1982-2016)", fontsize = 14, weight = "bold")
sns.despine()
plt.show()
```
